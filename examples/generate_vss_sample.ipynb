```python
from IPython.display import Image, display, HTML
import base64
import os

os.chdir('/content/visual-cryptography-cipherlab/examples/')

def create_image_div(filename, caption):
    """画像とキャプションのHTMLを生成"""
    with open(filename, 'rb') as f:
        img_data = base64.b64encode(f.read()).decode()

    return f"""
    <div style="flex:1; text-align:center; padding:10px; min-width:200px;">
        <img src="data:image/png;base64,{img_data}" style="width:100%; max-width:250px;">
        <p style="font-weight:bold; margin-top:10px;">{caption}</p>
    </div>
    """

# Flexboxで横並び
html = """
<div style="display:flex; flex-wrap:wrap; justify-content:space-around; align-items:flex-start;">
""" + \
create_image_div("secret.png", "元の秘密画像") + \
create_image_div("shareA.png", "シェアA") + \
create_image_div("shareB.png", "シェアB") + \
create_image_div("overlay.png", "A + B の重ね合わせ") + \
"""
</div>
"""

display(HTML(html))
```
